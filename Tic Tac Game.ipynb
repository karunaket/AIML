```python
def check_column(board, n, column):
    for i in range(n):
        if board[i][column] == 'Q':
            return False
    return True

def check_diagonal(board, n, row, column):
    # Check the upper left diagonal
    for i, j in zip(range(row, -1, -1), range(column, -1, -1)):
        if board[i][j] == 'Q':
            return False

    # Check the upper right diagonal
    for i, j in zip(range(row, -1, -1), range(column, n)):
        if board[i][j] == 'Q':
            return False

    return True

def is_safe(board, n, row, column):
    return check_column(board, n, column) and check_diagonal(board, n, row, column)

def nqn(board, row, n):
    if row == n:
        return True

    for col in range(n):
        if is_safe(board, n, row, col):
            board[row][col] = 'Q'
            if nqn(board, row + 1, n):
                return True
            board[row][col] = '.'  # backtrack

    return False

def solve_n_queens(n):
    board = [['.' for _ in range(n)] for _ in range(n)]
    if nqn(board, 0, n):
        for row in board:
            # print(' '.join(row))
            print(row)
    else:
        print("No solution exists")

# Input
n = int(input("Enter the length of Board: "))
solve_n_queens(n)
```

```python
import random

def board():
    """Creates an empty Tic-Tac-Toe board."""
    return [' ' for _ in range(9)]

def display(b):
    """Prints the current game board."""
    for i in range(3):
        print('|', *b[3 * i:3 * i + 3], '|')

def move(b, player, m):
    """Places the player's mark (X or O) on the given move."""
    b[m] = player

def win(b, p):
    """Checks if the given player has a winning streak."""
    win_conditions = ((0, 1, 2), (3, 4, 5), (6, 7, 8),
                      (0, 3, 6), (1, 4, 7), (2, 5, 8),
                      (0, 4, 8), (2, 4, 6))
    return any(all(b[i] == p for i in row) for row in win_conditions)

def full(b):
    """Checks if all spaces on the board are occupied."""
    return all(x != ' ' for x in b)

def get_player_move(b):
    """Gets the player's move."""
    while True:
        try:
            move = int(input("Move? (1-9) ")) - 1
            if 0 <= move < 9 and b[move] == ' ':
                return move
            print("Invalid move. Try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def random_move(b):
    """Makes a simple computer move (randomly choosing an empty space)."""
    return random.choice([i for i in range(9) if b[i] == ' '])

def play():
    """Starts and runs the Tic-Tac-Toe game."""
    b = board()
    player = 'X'  # Start with X playing first

    while not win(b, player) and not full(b):
        display(b)
        move(b, player, get_player_move(b) if player == 'X' else random_move(b))
        player = 'O' if player == 'X' else 'X'

    display(b)
    if win(b, player):
        print(f"Player {player} wins!")
    elif full(b):
        print("It's a tie!")

if __name__ == "__main__":
    play()

```
